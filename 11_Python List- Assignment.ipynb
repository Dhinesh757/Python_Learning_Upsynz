 **4. Write a Python program to remove duplicate elements from a list without using built-in functions like set().** 

```python
numbers = [2,3,3,4,5,6,5]
new_list = []
for i in numbers:
    if i not in new_list:
        new_list.append(i)
print(new_list)
```
